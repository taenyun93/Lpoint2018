# Association Rule

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#import modules
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

pd.set_option('display.max_columns', 100)
```

```python
df_interaction_full = pd.read_csv('Data/df_interaction_full.csv')
```

```python
df_interaction_full.CLNT_ID == df_interaction_full.CLNT_ID.astype('str')
df_interaction_full.PD_C == df_interaction_full.PD_C.astype('str')
df_interaction_full.CLAC3_NM == df_interaction_full.CLAC3_NM.astype('str')
```

```python
rand_inx = np.random.choice(np.unique(df_interaction_full[df_interaction_full.ACTION == 'product'].CLNT_ID),round(len(np.unique(df_interaction_full[df_interaction_full.ACTION == 'product'].CLNT_ID))*0.2),replace= False)
```

```python
basket = (df_interaction_full[df_interaction_full.ACTION == 'product'].set_index('CLNT_ID',drop = False).loc[rand_inx,:].reset_index(drop=True).groupby(['CLNT_ID','CLAC3_NM'])['PD_BUY_CT'].sum().unstack().reset_index().fillna(0).set_index('CLNT_ID'))
```

```python
basket.shape
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

## 장바구니 분석

basket_set 읽어들이기

```python
basket_sets= pd.read_pickle('Data/basket_set.pkl')
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.005, use_colnames=True)
```

```python
frequent_itemsets.head()
```

```python
len(frequent_itemsets)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
len(rules)
```

```python
rules.sort_values(by='lift',ascending = False)
```
